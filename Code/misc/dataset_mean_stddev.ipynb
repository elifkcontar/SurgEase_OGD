```
import numpy as np
from math import sqrt
from datasets.UCEIS_dataset import UCEIS_inmemory
```

```
frame_root = "/content/drive/My Drive/IBD/data_frames_and_videos"
annotation_root = "/content/drive/My Drive/IBD/data_03112020_unzipped"
all_folders = os.listdir(frame_root)

all_dataset = UCEIS_inmemory(all_folders, frame_root, annotation_root)
```

```
print(len(all_folders))
print(len(all_dataset))
```

```
R_total = 0
G_total = 0
B_total = 0

total_count = 0
for id, (image, _) in enumerate(all_dataset):
    if id%100==0:
        print(id)
            
    image = np.asarray(image)
    total_count = total_count + image.shape[0] * image.shape[1]

    R_total = R_total + np.sum(image[:, :, 0])
    G_total = G_total + np.sum(image[:, :, 1])
    B_total = B_total + np.sum(image[:, :, 2])

R_mean = R_total / total_count
G_mean = G_total / total_count
B_mean = B_total / total_count

R_total = 0
G_total = 0
B_total = 0

total_count = 0
for id, (image, _) in enumerate(all_dataset):
    if id%100==0:
        print(id)

    image = np.asarray(image)
    total_count = total_count + image.shape[0] * image.shape[1]

    R_total = R_total + np.sum((image[:, :, 0] - R_mean) ** 2)
    G_total = G_total + np.sum((image[:, :, 1] - G_mean) ** 2)
    B_total = B_total + np.sum((image[:, :, 2] - B_mean) ** 2)

R_std = sqrt(R_total / total_count)
G_std = sqrt(G_total / total_count)
B_std = sqrt(B_total / total_count)

print("R-mean: {:6.3f}, G-mean: {:6.3f}, B-mean: {:6.3f}".format(R_mean/255, G_mean/255, B_mean/255))
print("R-std: {:6.3f}, G-std: {:6.3f}, B-std: {:6.3f}".format(R_std/255, G_std/255, B_std/255))
```
