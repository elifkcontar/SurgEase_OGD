```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from utils.detect_blob import DetectBlob
```

```python
def show_image_PIL(image):
    image = np.array(image)
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

```python
original_folder_path = "../IBD/data_classified_scaled_320x568"
image_id = 100
```

```python
counter=1
for item in os.scandir(original_folder_path):
    if item.is_dir() and (not item.name.startswith(".")):
        print(str(counter)+" -> "+item.path)
        counter +=1
        
        image_path = os.path.join(item.path, "50.jpg")       
        im_original_pil = Image.open(image_path)
        
        show_image_PIL(im_original_pil)
        
        # With class
        detector = DetectBlob(min_threshold=10, min_area=60000, radius_scale=1.0)
        result = detector(im_original_pil)
        show_image_PIL(result)
```
